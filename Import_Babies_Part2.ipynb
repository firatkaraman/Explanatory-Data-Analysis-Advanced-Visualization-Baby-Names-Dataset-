## merging all files

```python
filenames
```

```python
dataframes = []
for name in filenames:
    df = pd.read_csv(name, header = None, names = ["State","Gender","Year", "Name", "Count"])
    dataframes.append(df)
```

```python
dataframes
```

```python
len(dataframes)
```

```python
df = pd.concat(dataframes, ignore_index=True)
```

```python
df
```

```python
df.info()
```

```python
df.to_csv("baby_names_state.csv", index=False)
```
